```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('yt_stat.csv')
df.head()
```

```python
df.describe()
```

```python
df.info
```

```python
df.info()
```

```python
#graph of top 10 channels with most subscribers
sns.barplot(x='subscribers', y='Youtuber', data=df[:10])
```

```python
#top 10 highest subscribers amount from countries
df2 = df.groupby('Country').sum().sort_values(by='subscribers', ascending=False)[:10]
df2
```

```python
#plot for df2
plt.figure(figsize=(10,5))
sns.barplot(x=df2.index, y='subscribers', data=df2)
```

```python
#biggest categories with most subscribers
df3 = df.groupby('category').sum().sort_values(by='subscribers', ascending=False)[:10]
df3
```

```python
#plot for df3
plt.figure(figsize=(20,10))
sns.barplot(x=df3.index, y='subscribers', data=df3)
```
